```python
actions = data.iloc[:, -495:-1].astype(np.int16)
print("actions: {}".format(actions))

```

```python
actions['action_list'] = actions.apply(lambda s: str([i for i in s.values]), axis=1)
print("actions['action_list']: {}".format(actions['action_list']))
print("actions['action_list']: {}".format(actions['action_list'].shape))
```

```python
print("actions.shape: {}".format(actions.shape))
nums = pd.value_counts(actions['action_list'])
print("nums: {}".format(nums.values))
print("nums_index: {}".format(nums.index))
print("nums_index_values: {}".format(nums.index.values))
```
